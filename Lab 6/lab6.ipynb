```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Problem 1

def odd_component(x):
    return 0.5 * (x - np.flip(x))

def even_component(x):
    return 0.5 * (x + np.flip(x))

# Define the time range
t = np.linspace(-10, 10, 1000)

# Define the signal x(t)
x = np.piecewise(t, [t < 0, t >= 0], [1, lambda t: np.sin(t)])

# Compute the odd and even components
odd = odd_component(x)
even = even_component(x)

# Plot the original signal and the odd and even components
plt.figure(figsize=(10, 6))
plt.subplot(3, 1, 1)
plt.plot(t, x)
plt.title('Original Signal x(t)')
plt.xlabel('t')
plt.ylabel('x(t)')

plt.subplot(3, 1, 2)
plt.plot(t, odd)
plt.title('Odd Component')
plt.xlabel('t')
plt.ylabel('Odd(t)')

plt.subplot(3, 1, 3)
plt.plot(t, even)
plt.title('Even Component')
plt.xlabel('t')
plt.ylabel('Even(t)')

plt.tight_layout()
plt.show()
```

```python
#Problem 2

# Define the values of t and frequencies
t_values = [0, 0.1, 0.2, 0.3, 0.4]
frequencies = [10, 15, 20]

# Create a two-dimensional array to store the results
results = np.zeros((len(frequencies), len(t_values)))

# Compute the values for all frequencies and store them in the array
for i, freq in enumerate(frequencies):
    for j, t in enumerate(t_values):
        x = 3 * np.cos(2 * np.pi * freq * t + 0.1)
        results[i, j] = x

# Print the results
for i, freq in enumerate(frequencies):
    print(f"Results for f = {freq} Hz:")
    for j, t in enumerate(t_values):
        print(f"x(t) = {results[i, j]} for t = {t} s")
    print()
```

```python
#Problem 3

def find_largest_t(omega):
    t = 0
    while True:
        expression1 = math.exp(1.2) * math.cos(omega * t)
        expression2 = t**3
        if expression1 < 10 and expression2 < 10:
            t += 0.01
        else:
            break
    return round(t - 0.01, 2)

omega_values = [35, 40, 45]
for omega in omega_values:
    largest_t = find_largest_t(omega)
    print(f"For omega = {omega}, largest t = {largest_t}")
```

```python
#Problem 4
def x(t):
    return (4 * np.cos(2 * np.pi * t + 0.2)) + (3 * np.sin((np.pi ** 2) * t))

# Create a 15-element vector with equally spaced intervals
t_values = np.linspace(0, 1, 15)
vector = x(t_values)

# Find the maximum and minimum element values
max_value = np.max(vector)
min_value = np.min(vector)

# Calculate the average of the element values
average_value = np.mean(vector)

# Find the indices of elements with magnitude greater than 4
indices = np.where(np.abs(vector) > 4)[0]

# Print the results
print("Maximum value:", max_value)
print("Minimum value:", min_value)
print("Average value:", average_value)
print("Indices with magnitude > 4:", indices)
```

```python
#Problem 5
f1 = 0.2
f2 = 0.425
t = np.arange(0, 10, 0.1)

# Define the functions
s1 = np.sin(f1 * t * 2 * np.pi)
s2 = 2 * np.pi * f2 * t + 0.4
s3 = s1 + s2

# Plot the functions
plt.plot(t, s1, label='s1 = sin(f1*t*2*pi)')
plt.plot(t, s2, label='s2 = (2*pi*f2*t + 0.4)')
plt.plot(t, s3, label='s3 = s1')

# Label the axes and create legends
plt.xlabel('t')
plt.ylabel('Value')
plt.legend()

# Show the plot
plt.show()
```

```python
#Problem 6

def MySinc(x):
    if x == 0:
        return 1
    else:
        return np.sin(x) / x

x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y_mysinc = np.vectorize(MySinc)(x)
y_matlabsinc = np.sinc(x / np.pi)

plt.plot(x, y_mysinc, label='MySinc()')
plt.plot(x, y_matlabsinc, label='MATLAB sinc()')
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.legend()
plt.title('Comparison of MySinc() and MATLAB sinc()')
plt.grid(True)
plt.show()
```
